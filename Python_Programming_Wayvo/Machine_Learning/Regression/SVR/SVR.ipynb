```python
import pandas as pd
```

```python
svdf=pd.read_csv('Housing.csv')
svdf
```

```python
x=svdf.drop('price',axis=1)
y=svdf['price']
```

```python
from sklearn.preprocessing import LabelEncoder
lbe = LabelEncoder()
# x['mainroad']= lbe.fit_transform(x['mainroad'])
# x['guestroom']= lbe.fit_transform(x['guestroom'])
# x['basement']= lbe.fit_transform(x['basement'])
# x['hotwaterheating']= lbe.fit_transform(x['hotwaterheating'])
# x['airconditioning']= lbe.fit_transform(x['airconditioning'])
# x['prefarea']= lbe.fit_transform(x['prefarea'])
# x['furnishingstatus']= lbe.fit_transform(x['furnishingstatus'])
x[['mainroad', 'guestroom', 'basement', 'hotwaterheating', 'airconditioning', 'prefarea', 'furnishingstatus']] = x[['mainroad', 'guestroom', 'basement', 'hotwaterheating', 'airconditioning', 'prefarea', 'furnishingstatus']].apply(lbe.fit_transform)
x
```

```python
from sklearn.preprocessing import StandardScaler
sa=StandardScaler()
sbe=StandardScaler()
sba=StandardScaler()
ss=StandardScaler()
sm=StandardScaler()
sg=StandardScaler()
sb=StandardScaler()
sp=StandardScaler()
shw=StandardScaler()
sac=StandardScaler()
spr=StandardScaler()
sf=StandardScaler()
x['area']=sa.fit_transform(x[['area']])
x['bedrooms']=sbe.fit_transform(x[['bedrooms']])
x['bathrooms']=sba.fit_transform(x[['bathrooms']])
x['stories']=ss.fit_transform(x[['stories']])
x['mainroad']=sm.fit_transform(x[['mainroad']])
x['guestroom']=sg.fit_transform(x[['guestroom']])
x['parking']=sp.fit_transform(x[['parking']])
x['basement']=sb.fit_transform(x[['basement']])
x['hotwaterheating']=shw.fit_transform(x[['hotwaterheating']])
x['airconditioning']=sac.fit_transform(x[['airconditioning']])
x['prefarea']=spr.fit_transform(x[['prefarea']])
x['furnishingstatus']=sf.fit_transform(x[['furnishingstatus']])

x
```

```python
ys=StandardScaler()
y=ys.fit_transform(y.values.reshape(-1,1))
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
from sklearn.svm import SVR
svr=SVR(kernel='rbf')
svr.fit(x_train,y_train)
```

```python
y_pred=svr.predict(x_test)
out=ys.inverse_transform(y_pred.reshape(-1,1))
out
```

```python
from sklearn.metrics import r2_score
accuracy_score=r2_score(y_test,y_pred)
accuracy_score
```
