```python
import pandas as pd
```

```python
from sklearn.datasets import load_iris
# ir=load_iris()
# ir=ir.drop(['sepal length (cm)','sepal width (cm)'])
# ir
ir = load_iris()
ir_df = pd.DataFrame(ir.data, columns=ir.feature_names)
ir_df = ir_df.drop(['sepal length (cm)', 'sepal width (cm)'], axis=1)
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
```

```python
km=KMeans(n_clusters=3)
```

```python
km_pred=km.fit_predict(ir_df)
km_pred
```

```python
km.cluster_centers_
```

```python
ir_df['cluster']=km_pred
ir_df
```

```python
df1 = ir_df[ir_df.cluster==0]
df2 = ir_df[ir_df.cluster==1]
df3 = ir_df[ir_df.cluster==2]
plt.scatter(df1['petal length (cm)'],df1['petal width (cm)'],color='green')
plt.scatter(df2['petal length (cm)'],df2['petal width (cm)'],color='red')
plt.scatter(df3['petal length (cm)'],df3['petal width (cm)'],color='black')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color='purple',marker='*',label='centroid')
plt.xlabel('Petal length (cm)')
plt.ylabel('Petal Width (cm)')
plt.legend()
```

```python
scaler=MinMaxScaler()
scaler.fit(ir_df[['petal width (cm)']])
ir_df['petal width (cm)'] = scaler.transform(ir_df[['petal width (cm)']])

scaler.fit(ir_df[['petal length (cm)']])
ir_df['petal length (cm)'] = scaler.transform(ir_df[['petal length (cm)']])

ir_df
```

```python
nkm=KMeans(n_clusters=3)
```

```python
nkm_pred=nkm.fit_predict(ir_df)
nkm_pred
```

```python
nkm.cluster_centers_
```
